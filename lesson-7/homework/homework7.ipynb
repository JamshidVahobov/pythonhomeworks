```python
def is_prime(n):
    for b in range(2,int(n**0.5)):
        if n % b == 0:
            return False
        return True
    
prime = is_prime(12)
prime
```

```python
def digit_sum(k):
    sum = 0
    k = str(k)
    for i in k:
        sum += int(i)
    print(sum)
digit_sum(5920)
```

```python
def two_level(n):
    n_list = []
    i = 2
    while i < n:
        n_list.append(i)
        i = i*2
    print(' '.join(map(str,n_list)))

two_level(33)
```
