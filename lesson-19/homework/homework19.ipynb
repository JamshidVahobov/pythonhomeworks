```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('task\\sales_data.csv')
df
```

```python
df.groupby('Category').agg({'Quantity':'sum','Price':'mean','Quantity':'max'})
```

```python
df.loc[df.groupby('Category')['Quantity'].idxmax(), ['Category', 'Product', 'Quantity']]
```

```python
df['Total'] = df['Quantity'] * df['Price']
df.groupby('Date').agg({'Total': 'max'}).max()
```
