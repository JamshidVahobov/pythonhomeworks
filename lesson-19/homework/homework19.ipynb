```python
import pandas as pd
import numpy as np
import sqlite3
#sqlite
connection = sqlite3.connect('task\population.db')

cursor = connection.cursor()

query = "SELECT * from population;"

results = cursor.execute(query)

row = results.fetchall()

cursor.close()
connection.commit()
connection.close()
```

```python
columns = ['id','first_name','last_name','email','gender','salary','state']
population = pd.DataFrame(row,columns=columns)
population_salary = pd.read_excel('task\\population_salary_analysis.xlsx')
```

```python
bins = [0,200000,400000,600000,800000,1000000,1200000,1400000,1600000,1800000,100000000]
labels = ['till $200,000','$200,001 - $400,000','$400,001 - $600,000','$600,001 - $800,000','$800,001 - $1,000,000','$1,000,001 - $1,200,000','$1,200,001 - $1,400,000','$1,400,001 - $1,600,000','$1,600,001 - $1,800,000','$1,800,001 and over']
population['Salary Band'] = pd.cut(population['salary'],bins=bins,labels=labels)
```

```python
population_grouped = population.groupby('Salary Band').agg(
    Percentage=pd.NamedAgg(column='id', aggfunc='count'),
    Average_Salary=pd.NamedAgg(column='salary', aggfunc='mean'),
    Median_Salary=pd.NamedAgg(column='salary', aggfunc='median'),
    Number_of_population=pd.NamedAgg(column='id', aggfunc='count')
)

# Calculate percentage
total_count = population_grouped['Percentage'].sum()
population_grouped['Percentage'] = (population_grouped['Percentage'] / total_count) * 100

population_grouped['Average_Salary'] = population_grouped['Average_Salary'].astype(int)
population_grouped['Median_Salary'] = population_grouped['Median_Salary'].astype(int)
population_grouped['Percentage'] = population_grouped['Percentage'].map("{:.2f}%".format)
population_grouped
```
