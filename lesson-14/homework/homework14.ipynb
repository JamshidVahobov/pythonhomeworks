```python
import json
import os

print("Welcome to books.json")

# Initialize the JSON file if it doesn't exist
if not os.path.exists("books.json"):
    with open("books.json", "w") as f:
        json.dump({}, f, indent=4)

while True:
    print("\nMain menu:")
    print("1 - Add a new book")
    print("2 - Update an existing book")
    print("3 - Delete a book")
    print("4 - Delete all books")
    print("5 - Exit the program")

    try:
        choice = int(input("Enter the action number: "))
    except ValueError:
        print("Please enter a valid number.")
        continue

    if choice == 1:
        book_name = input("Enter the book name: ")
        book_info = input("Enter the information about the book: ")

        # Read existing data safely
        try:
            with open("books.json", "r") as file:
                content = file.read()
                data = json.loads(content) if content else {}
        except json.JSONDecodeError:
            print("⚠️ Warning: books.json was corrupted or not in JSON format. Resetting it.")
            data = {}

        # Add/update the book
        data[book_name] = book_info

        # Save it back
        with open("books.json", "w") as file:
            json.dump(data, file, indent=4)

        print(f"✅ '{book_name}' has been saved to books.json.")

    elif choice == 5:
        print("👋 Exiting program.")
        break

    else:
        print("🚧 This option is not implemented yet.")
```
