```python
import json
import os

print("Welcome to books.json")

if not os.path.exists("books.json"):
    with open("books.json", "w") as f:
        json.dump({}, f, indent=4)

while True:
    print("\nMain menu:")
    print("1 - Add a new book")
    print("2 - Update an existing book")
    print("3 - Delete a book")
    print("4 - Delete all books")
    print("5 - Exit the program")

    try:
        choice = int(input("Enter the action number: "))
    except ValueError:
        print("❌ Please enter a valid number.")
        continue

    try:
        with open("books.json", "r") as file:
            content = file.read()
            data = json.loads(content) if content else {}
    except json.JSONDecodeError:
        print("⚠️ Warning: books.json was corrupted or not in JSON format. Resetting it.")
        data = {}

    if choice == 1:
        book_name = input("Enter the book name: ")
        book_info = input("Enter the information about the book: ")
        data[book_name] = book_info

        with open("books.json", "w") as file:
            json.dump(data, file, indent=4)

        print(f"✅ '{book_name}' has been saved to books.json.")

    elif choice == 2:
        print("📚 Existing books:")
        for name in data:
            print(f"- {name}")
        book_name = input("Enter the book name to update: ")
        if book_name in data:
            book_info = input("Enter the new information: ")
            data[book_name] = book_info
            with open("books.json", "w") as file:
                json.dump(data, file, indent=4)
            print(f"🔄 '{book_name}' has been updated.")
        else:
            print("❌ Book not found.")

    elif choice == 3:
        print("📚 Existing books:")
        for name in data:
            print(f"- {name}")
        book_name = input("Enter the book name to delete: ")
        if book_name in data:
            data.pop(book_name)  # ✅ Fix: use pop, not popitem
            with open("books.json", "w") as file:
                json.dump(data, file, indent=4)
            print(f"🗑️ '{book_name}' has been deleted.")
        else:
            print("❌ Book not found.")

    elif choice == 4:
        confirm = input("Are you sure you want to delete ALL books? (yes/no): ").lower()
        if confirm == "yes":
            data = {}  # reset data
            with open("books.json", "w") as file:
                json.dump(data, file, indent=4)
            print("🧹 All books have been deleted.")
        else:
            print("Cancelled.")

    elif choice == 5:
        print("👋 Exiting program.")
        break

    else:
        print("🚧 This option is not implemented yet.")
```

```python
import requests
import random

API_KEY = "e8647089"
OMDB_URL = "http://www.omdbapi.com/"

movie_titles = [
    "Inception", "The Dark Knight", "Titanic", "The Godfather",
    "Forrest Gump", "Pulp Fiction", "The Matrix", "The Shawshank Redemption",
    "Avengers: Endgame", "Interstellar", "The Silence of the Lambs"
]

def get_movie_info(title):
    params = {
        "t": title,
        "apikey": API_KEY
    }
    response = requests.get(OMDB_URL, params=params)
    return response.json()

def recommend_movie_by_genre(genre):
    matching_movies = []

    for title in movie_titles:
        movie = get_movie_info(title)
        if movie.get("Response") == "True" and genre.lower() in movie.get("Genre", "").lower():
            matching_movies.append(movie)

    if not matching_movies:
        print(f"Sorry, no movies found for genre: {genre}")
    else:
        recommended = random.choice(matching_movies)
        print(f"\n🎬 Movie Recommendation:\nTitle: {recommended['Title']}\nYear: {recommended['Year']}\nGenre: {recommended['Genre']}\nPlot: {recommended['Plot']}")

user_genre = input("Enter a movie genre (e.g., Action, Comedy, Drama): ")
recommend_movie_by_genre(user_genre)
```
